```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
hotel=pd.read_csv("C:/Users/Vivek/OneDrive/Desktop/hotel_booking.csv")
hotel
```

```python
hotel.head()
```

```python
hotel.isnull().sum()
```

```python
hotel.drop(["agent","company"],axis=1,inplace=True)
```

```python
hotel.isna().sum()
```

```python
hotel.dropna(inplace=True)
```

```python
hotel.info()
```

```python
hotel.describe()
```

```python
hotel.columns
```

```python
hotel["reservation_status_date"]
```

```python
hotel.rename(columns={"is_canceled":"canceled"},inplace=True)
```

```python
hotel["canceled"].replace(1,"Yes",inplace=True)
```

```python
hotel["canceled"].replace(0,"No",inplace=True)
```

```python
cancel_status=hotel['canceled'].value_counts()
cancel_status
```

```python
hotel.shape
```

```python
plt.figure(figsize=(5,4))
plt.title("Reservation_Status")
x=sns.countplot(x='canceled',data=hotel)
x
for bars in x.containers:
    x.bar_label(bars)
```

```python
plt.title("Reservation_for_different_hotels")
y=sns.countplot(x="hotel",hue="canceled",data=hotel)
plt.ylabel("No_of_Reservation")
y
for bars in y.containers:
    y.bar_label(bars)

```

```python
plt.figure(figsize=(12,5))
plt.title("Reservation_Status_per_month")
m=sns.countplot(x="arrival_date_month",hue="canceled",data=hotel)
m
for bars in m.containers:
    m.bar_label(bars)
```

```python
data=hotel[hotel["canceled"]=="Yes"].groupby(["arrival_date_month"],as_index=False)["adr"].mean()
data
```

```python
plt.figure(figsize=(12,4))
sns.barplot(x="arrival_date_month",y="adr",data=data)
plt.title("ADR_per_month")
plt.show()
```

```python
m=hotel[hotel["canceled"]=="Yes"].groupby(["arrival_date_month"],as_index=False)["adr"].mean()
m
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x="arrival_date_month",y="adr",data=m)
plt.show()
```

```python
canceled_data=hotel[hotel["canceled"]=="Yes"]
top_10_country=canceled_data["country"].value_counts()[ :10]
plt.pie(top_10_country,labels=top_10_country.index,autopct="%.1f")
plt.title("Top_10_country_with_highest_cancelation_rate")
plt.show()
```
